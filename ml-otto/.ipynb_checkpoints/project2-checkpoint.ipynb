```python

from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
from scipy import stats
from sklearn.ensemble import VotingClassifier
import pandas as pd
import numpy
```

```python
# load data
test_dat = loadtxt('test_sample_proj2.csv', delimiter=",",skiprows=1)
train_dat = loadtxt('train_sample_proj2.csv', delimiter=",",skiprows=1)
```

```python
train_dat[1,]
```

```python
test_dat[0:3,]
```

```python

# split data into X and y
X = train_dat[:,0:92]
Y = train_dat[:,93]
seed = 7
test_size = 0.20
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
y_test[2,]
```

```python
stats.itemfreq(Y)
```

```python


# fit model no training data
model = XGBClassifier(max_depth=10,n_estimators=400,objective='multi:softprob',gamma=0,learning_rate=0.1,
                     min_child_weight=1,subsample=0.9,colsample_bytree=0.8,nthread=16)
# max_delta_step=1
eval_set = [(X_test, y_test)]
model.fit(X_train, y_train,eval_metric = 'mlogloss',eval_set=eval_set,verbose=True,early_stopping_rounds=100,
          xgb_model='model1')
#,
print(model.best_iteration)
best_iter = model.best_iteration

# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]



# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
model.best_iteration
```
