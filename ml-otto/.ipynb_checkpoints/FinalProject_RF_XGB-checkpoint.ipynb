```python
def modelfit(alg, dtrain, predictors,useTrainCV=True, cv_folds=5, early_stopping_rounds=50):
    xVals = np.log(1 + dtrain[predictors])
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(xVals, label=yTrain)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='mlogloss', early_stopping_rounds=early_stopping_rounds, verbose_eval=True)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    mdl = alg.fit(xVals, yTrain,eval_metric='logloss')
        
    #Predict training set:
    #dtrain_predictions = alg.predict(xVals)
    #dtrain_predprob = alg.predict_proba(xVals)
        
    #Print model report:
    #print("\nModel Report")
    #print("Accuracy : %.4g" % metrics.accuracy_score(yTrain, dtrain_predictions))
    #print("AUC Score (Train): %f" % metrics.roc_auc_score(yTrain, dtrain_predprob))
                    
    #feat_imp = pd.Series(alg.booster().get_fscore()).sort_values(ascending=False)
    #feat_imp.plot(kind='bar', title='Feature Importances')
    #plt.ylabel('Feature Importance Score')
    
    return mdl
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.utils import to_categorical, plot_model
from keras.optimizers import Adam,SGD,Adagrad
from keras.utils.vis_utils import model_to_dot
from keras.wrappers.scikit_learn import KerasClassifier
from IPython.display import SVG
from sklearn.ensemble import VotingClassifier
from sklearn import model_selection
```

```python
num_features = xTrain.shape[1]

encoder = LabelEncoder()
yNeural = encoder.fit_transform(yTrain).astype(np.int32)
yCompleteNeural = encoder.fit_transform(y).astype(np.int32)

num_classes = len(encoder.classes_)

# convert class vectors to binary class matrices
yNeural = to_categorical(yNeural, num_classes)
yCompleteNeural = to_categorical(yCompleteNeural, num_classes)
print(yNeural)
print(yTrain)
```

```python
def getModel(dropout=0.1, neurons1=500, neurons2=250,
             learningRate=0.004):
    np.random.seed(1)
    
    model = Sequential()
    
    model.add(Dense(neurons1, activation='relu', input_dim=num_features,
                    name='Dense_1'))
    model.add(Dropout(dropout,name='Dropout_1'))
    
    model.add(Dense(neurons2, activation='relu',name='Dense_2'))
    model.add(Dropout(dropout,name='Dropout_2'))
    
    
    #model.add(Dense(10, activation='relu',name='Dense_3'))
    #model.add(Dropout(0.2,name='Dropout_3'))
    
    model.add(Dense(num_classes, activation='softmax',name='Output'))
    
    model.compile(loss='categorical_crossentropy',
                  optimizer=Adam(lr=learningRate), metrics=['accuracy'])
    return model

#SVG(model_to_dot(model1).create(prog='dot', format='svg'))
```

```python
def getModel_2(dropout=[0.13, 0.77, 0.35, 0.1,], neurons=[2500, 1300, 40],
             learningRate=0.04):
    
    np.random.seed(1)
    model = Sequential()
    
    model.add(Dense(neurons[0], activation='relu', input_dim=num_features, name='Dense_1'))
    model.add(Dropout(dropout[0],name='Dropout_1'))
    
    model.add(Dense(neurons[1], activation='relu',name='Dense_2'))
    model.add(Dropout(dropout[1],name='Dropout_2'))
    
    model.add(Dense(neurons[2], activation='relu',name='Dense_3'))
    model.add(Dropout(dropout[2],name='Dropout_3'))
    
    model.add(Dense(num_classes, activation='softmax',name='Output'))
    model.add(Dropout(dropout[3],name='Dropout_4'))
        
    model.compile(loss='categorical_crossentropy',
                  optimizer=Adagrad(lr=learningRate, decay=(learningRate/600)), metrics=['accuracy'])
    return model

```

```python
def getModelLevel2(dropout=0.1, neurons1=500, neurons2=250,
             learningRate=0.004, num_features=1):
    np.random.seed(1)
    
    model = Sequential()
    
    model.add(Dense(neurons1, activation='linear', input_dim=num_features,
                    name='Dense_1'))
    model.add(Dropout(dropout,name='Dropout_1'))
    
    #model.add(Dense(neurons2, activation='relu',name='Dense_2'))
    #model.add(Dropout(dropout,name='Dropout_2'))
    
    
    #model.add(Dense(40, activation='relu',name='Dense_3'))
    #model.add(Dropout(0.1,name='Dropout_3'))
    
    model.add(Dense(num_classes, activation='softmax',name='Output'))
    
    model.compile(loss='categorical_crossentropy',
                  optimizer=Adam(lr=learningRate), metrics=['accuracy'])
    return model

#SVG(model_to_dot(model1).create(prog='dot', format='svg'))
```

## Support Vector Machines (SVM)

```python
from sklearn.svm import SVC

kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)

def svc_param_selection(X, y, nfolds):
    Cs = [0.001, 0.01, 0.1, 1, 10]
    gammas = [0.001, 0.01, 0.1, 1]
    param_grid = {'C': Cs, 'gamma' : gammas}
    grid_search = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=nfolds, verbose=2)
    grid_search.fit(X, y)
    grid_search.best_params_
    return grid_search.best_params_

svc_param_selection(preprocessing.scale(xLog), yTrain, 5)
```
