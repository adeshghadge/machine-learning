```python

def writeFileNamesInTextFile(dirPath, textFileName, folderName):
    with open(dirPath + textFileName, "w") as a:
        for path, subdirs, files in os.walk(dirPath + folderName):
            for filename in files:
                if('.png' not in filename):
                    continue
                a.write(str(filename).replace('.png', '') + os.linesep) 
            

def writeFilePathsInTextFile(dirPath, textFileName, folderName):
    with open(dirPath + textFileName, "w") as a:
        for path, subdirs, files in os.walk(dirPath + folderName):
            for filename in files:
                if('.png' not in filename):
                    continue
                f = os.path.join(path, filename)
                a.write(str(f) + os.linesep) 
                
```

```python
import numpy as np
import h5py
import pickle

def getImageProperties(imgDir, fileName):

    f = h5py.File('%s/%s.mat'%(imgDir, fileName))

    metadata= {}
    metadata['height'] = []
    metadata['label'] = []
    metadata['left'] = []
    metadata['top'] = []
    metadata['width'] = []

    def print_attrs(name, obj):
        vals = []
        if obj.shape[0] == 1:
            vals.append(int(obj[0][0]))
        else:
            for k in range(obj.shape[0]):
                vals.append(int(f[obj[k][0]][0][0]))
        metadata[name].append(vals)

    for item in f['/%s/bbox'%(fileName)]:
        f[item[0]].visititems(print_attrs)

    with open(fileName+'metadata.pickle','wb') as pf:
      pickle.dump(metadata, pf, pickle.HIGHEST_PROTOCOL) 
    
    return metadata
```

```python
def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```

```python
from PIL import Image, ImageDraw, ImageFont

def convert_annotation(metadata, img, image_set, image_id):

    out_file = open('%slabels/%s_%s/%s.txt'%(projDir, img, image_set, image_id), 'w')

    im = Image.open('%s%s.png'%(trainDir, str(image_id)))
    w, h = im.size

    imgIndex = int(image_id) - 1
    
    for j in range(0, len(metadata['height'][imgIndex])):
        cls = str(metadata['label'][imgIndex][j])
        if(cls == '10'):
            cls = '0'
        
        cls_id = classes.index(cls)

        b = (float(metadata['left'][imgIndex][j]), float(metadata['left'][imgIndex][j]+metadata['width'][imgIndex][j]), \
             float(metadata['top'][imgIndex][j]), float(metadata['top'][imgIndex][j]+metadata['height'][imgIndex][j]))
        bb = convert((w,h), b)

        out_file.write(str(cls_id) + " " + " ".join([str(a) for a in bb]) + '\n')
```

```python
def convert_annotation(img, image_set, image_id):
    in_file = open('VOCdevkit/VOC%s/Annotations/%s.xml'%(img, image_id))
    out_file = open(projDir+'labels/%s/%s.txt'%(image_set, image_id), 'w')
    tree=ET.parse(in_file)
    root = tree.getroot()
    size = root.find('size')
    
    w = int(size.find('width').text)
    h = int(size.find('height').text)

    for obj in root.iter('object'):
        difficult = obj.find('difficult').text
        cls = obj.find('name').text
        if cls not in classes or int(difficult) == 1:
            continue
        cls_id = classes.index(cls)
        xmlbox = obj.find('bndbox')
        b = (float(xmlbox.find('xmin').text), float(xmlbox.find('xmax').text), float(xmlbox.find('ymin').text), float(xmlbox.find('ymax').text))
        bb = convert((w,h), b)
        out_file.write(str(cls_id) + " " + " ".join([str(a) for a in bb]) + '\n')
```
